# Employee Turnover Predictor

Employee attrition is a commonly sought after information or tool ofthat could help human resource of various companies minimize the time, effort, and money in hiring employee(s). With such a predictor not only can companies minimized resource loses on training new recruits but also determining the type of people and background the company is actually seeking to fit the vacant posistion. According to the U.S. Bureau of Statistics in 2018, the average turnover rate in the U.S. is about 12-15% annually. While LinkedIn shows an average annual worldwide employee turnover rate of 10.9% in 2018. This is a significant number that companies should pay attention to since 1 in every 10 recruits they hire has a potential of turning over. The amount of resource that a company places into a new recruit is also no small anmount. Therefore, in this notebook I have attempted to create a few employee turnover predictors.

## 6. Conclusion

Out of all the algorithms used, SVM showed the lowest ROC-AUC scores. Although I have selected the RF and GB as my method of modeling due to its high ROC-AUC scores, I know that this may not neccessarily by the best solution as the model could be overfitting.<br><br>

The used of Cross Validation in this notebook is to prevent overfitting of model on the same dataset. This method splits training set into k-smaller sets where the models are trained using k-number of folds and predictions are validated on remaining testing sets.<br><br>

Noticed that the hyperparameters of the GB algorithm was not too fine-tuned. This is because my current setup does not allow me to perform too many computations to achieve optimized hyperparameters.<br><br>
